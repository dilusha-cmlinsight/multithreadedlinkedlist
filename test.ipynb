```python
import pandas as pd
```

```python
case_1 = pd.read_csv('output/Case_1.csv')
case_2 = pd.read_csv('output/Case_2.csv')
case_3 = pd.read_csv('output/Case_3.csv')
```

```python
case_1.head()
```

```python
metrics = ['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']
```

```python
case_1['Serial'].mean()
```

```python
case_1_data = {}

for col in case_1.columns:
    if not col in case_1_data.keys():
        d = case_1[col].describe()
        case_1_data[col] = {}
        for m in metrics:
            v = d[m]
            case_1_data[col][m] = v

stats_case_1 = pd.DataFrame(data = case_1_data, columns = case_1.columns, index = metrics)

stats_case_1
```
